```python
import pandas as pd # pandas is used to load and manipulate data and for One-Hot Encoding
import numpy as np # numpy is used to calculate the mean and standard deviation
import matplotlib.pyplot as plt # matplotlib is for drawing graphs
import matplotlib.colors as colors
from sklearn.model_selection import train_test_split # split  data into training and testing sets
from sklearn.preprocessing import scale # scale and center data
from sklearn.svm import SVC # this will make a support vector machine for classificaiton
from sklearn.model_selection import GridSearchCV # this will do cross validation and hyperparameter tuning
from sklearn.metrics import confusion_matrix # this creates a confusion matrix
from sklearn.metrics import classification_report

```

```python
df=pd.read_csv("processed.cleveland.data", header=None)
```

```python
df.columns = ['age',
              'sex',
              'cp',
              'restbp',
              'chol',
              'fbs',
              'restecg',
              'thalach',
              'exang',
              'oldpeak',
              'slope',
              'ca',
              'thal',
              'hd']
df.head()
```

```python
#we will drop the rows which have ?
df_final=df.loc[(df['ca'] != '?') & (df['thal'] != '?')]
```

```python
df_final
```

```python
X = df_final.drop('hd', axis=1)
Y = df_final['hd']
```

```python
#For the cp (chest pain) column, we have 4 options:

# typical angina,
# atypical angina,
# non-anginal pain,
# asymptomatic
```

```python
print(X['cp'].unique())
print(X['restecg'].unique())
print(X['slope'].unique())
print(X['thal'].unique())
```

```python
X_encoded = pd.get_dummies(X, columns=['cp', 'restecg', 'slope', 'thal'])
X_encoded.head()
```

```python
#Now, one last thing before we build a Support Vector Machine. y doesn't just contain 0s and 1s. Instead, it has 5 different levels of heart disease.
# 0 = no heart disease and 1-4 are various degrees of heart disease
```

```python
Y.unique()
```

```python
Y = np.where(Y > 0, 1, Y)
```

```python
np.unique(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = 42)
X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

```python
clf_svm = SVC(random_state=42)
clf_svm.fit(X_train_scaled, y_train)
```

```python
y_pred = clf_svm.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```

```python
#Hyperparameter Tuning
```

```python
#here it is performing a grid search using cross validation to find the optimal hyperparameter for a support vector classifier
param_grid = [{'C': [1, 10, 100, 1000],
               'gamma': [0.001, 0.0001],
               'kernel': ['rbf']}]

optimal_params = GridSearchCV(SVC(), param_grid, cv=5, verbose=0)

optimal_params.fit(X_train_scaled, y_train)
optimal_params.best_params_
```

```python
clf_svm = SVC(random_state=42, C=10, gamma =0.001, kernel='rbf')
clf_svm.fit(X_train_scaled, y_train)
```

```python
y_pred = clf_svm.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```
